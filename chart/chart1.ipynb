```python
import numpy as np
import matplotlib.pyplot as plt

def highlight_numbers_in_table(numbers_to_highlight):
    # 创建一个6x6的矩阵并填充数字
    table = np.arange(1, 37).reshape(6, 6)

    # 绘制表格
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_facecolor('white')  # 设置背景色为白色

    # 绘制背景颜色为红色的矩形框
    for i in range(table.shape[0]):
        for j in range(table.shape[1]):
            if table[i, j] in numbers_to_highlight:
                ax.add_patch(plt.Rectangle((j-0.5, i-0.5), 1, 1, fill=True, color='red', edgecolor='grey'))

    # 添加文字
    for i in range(table.shape[0]):
        for j in range(table.shape[1]):
            ax.text(j, i, str(table[i, j]), ha='center', va='center', color='black')

    # 设置网格线
    ax.set_xticks(np.arange(-0.5, 6, 1))
    ax.set_yticks(np.arange(-0.5, 6, 1))
    ax.grid(which='both', color='grey', linestyle='-', linewidth=1)
    
    # 隐藏刻度
    ax.tick_params(which='both', bottom=False, left=False, labelbottom=False, labelleft=False)

    plt.show()

# 示例调用
highlight_numbers_in_table([3, 8, 15, 22, 27, 35])
```
